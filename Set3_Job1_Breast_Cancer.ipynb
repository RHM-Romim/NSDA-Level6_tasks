## Task 8: Load the Breast Cancer Dataset into the environment.

```python
import pandas as pd

# Load the dataset, dropping the 'Unnamed: 0' column which is likely an index artifact
df = pd.read_csv('/content/Group 2 Job 1_2_3 _ Breast Cancer Dataset (1).csv', index_col=0)

print("Dataset loaded successfully, 'Unnamed: 0' column removed.")
```

## Task 9: Perform initial data inspection (shape, data types, missing values, summary statistics).

```python
# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
display(df.head())

# Get the shape of the dataset
print("\nShape of the dataset (rows, columns):")
print(df.shape)

# Get data types of each column
print("\nData types of each column:")
print(df.info())

# Check for missing values
print("\nMissing values in each column:")
display(df.isnull().sum())

# Generate summary statistics for numerical columns
print("\nSummary statistics for numerical columns:")
display(df.describe())
```

## Task 10: Compute mean, median, mode, variance, and standard deviation for all numerical features.

```python
# Select only numerical columns for calculations
numerical_cols = df.select_dtypes(include=['number']).columns

# Compute mean
print("\nMean for numerical features:")
display(df[numerical_cols].mean())

# Compute median
print("\nMedian for numerical features:")
display(df[numerical_cols].median())

# Compute mode (can return multiple values, so handling as a series of lists)
print("\nMode for numerical features:")
for col in numerical_cols:
    print(f"  {col}: {df[col].mode().tolist()}")

# Compute variance
print("\nVariance for numerical features:")
display(df[numerical_cols].var())

# Compute standard deviation
print("\nStandard deviation for numerical features:")
display(df[numerical_cols].std())
```

## Task 11: Compute covariance between selected variable pairs and interpret the results.

```python
# Select a few numerical features for covariance calculation
# Note: The 'Unnamed: 0' column, which appears to be an unwanted index column,
# is automatically excluded by this selection of specific features.
# For cleaner data, it's generally recommended to drop such columns during initial data loading, e.g., df = pd.read_csv('file.csv', index_col=0)
selected_features = ['x.radius_mean', 'x.texture_mean', 'x.area_mean', 'x.perimeter_mean']

# Compute the covariance matrix
covariance_matrix = df[selected_features].cov()

print("\nCovariance Matrix for selected features:")
display(covariance_matrix)

print("\nInterpretation of Covariance:\nPositive covariance indicates that the variables tend to move in the same direction. Negative covariance means they tend to move in opposite directions. A covariance close to zero suggests no linear relationship. The magnitude of the covariance is not standardized, so it's best for understanding direction rather than strength.")
print("\nFrom the covariance matrix:")
print("  - 'x.radius_mean' and 'x.area_mean' show a very high positive covariance (approx. 1224), indicating a strong tendency to increase together. This is expected as a larger radius typically results in a larger area.")
print("  - Similarly, 'x.radius_mean' and 'x.perimeter_mean' also have a high positive covariance (approx. 85), which is logical given their geometrical relationship.")
print("  - 'x.texture_mean' has relatively lower covariance values with other features compared to the high inter-correlations among radius, area, and perimeter, suggesting it might behave more independently or have weaker linear relationships with these physical dimensions.")
print("  - All shown covariances are positive, implying that as one of these features increases, the others tend to increase as well, reflecting the overall growth of a tumor.")
```

## Task 12: Perform hypothesis testing (t-test) to compare two selected groups in the dataset.

```python
from scipy import stats

# Assuming 'y' is the categorical column (M=Malignant, B=Benign)
# We will compare 'x.radius_mean' between Malignant and Benign groups

# Separate data into two groups based on the 'y' column
group_m = df[df['y'] == 'M']['x.radius_mean']
group_b = df[df['y'] == 'B']['x.radius_mean']

# Perform independent two-sample t-test
t_stat, p_value = stats.ttest_ind(group_m, group_b, equal_var=False) # Assuming unequal variances

print(f"\nHypothesis Testing (t-test) for x.radius_mean between Malignant and Benign groups:")
print(f"  T-statistic: {t_stat:.4f}")
print(f"  P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print(f"  Since p-value ({p_value:.4f}) < alpha ({alpha}), we reject the null hypothesis.")
    print("  There is a statistically significant difference in mean 'x.radius_mean' between Malignant and Benign tumors.")
else:
    print(f"  Since p-value ({p_value:.4f}) >= alpha ({alpha}), we fail to reject the null hypothesis.")
    print("  There is no statistically significant difference in mean 'x.radius_mean' between Malignant and Benign tumors.")
```

## Task 13: Calculate confidence intervals (95%) for selected features.

```python
# Calculate 95% confidence interval for 'x.area_mean'

confidence_level = 0.95
data = df['x.area_mean']

# Calculate sample mean and standard error of the mean
mean_data = data.mean()
std_err_data = stats.sem(data)

# Calculate margin of error using t-distribution (since population std dev is unknown)
# Degrees of freedom = n - 1
degrees_freedom = len(data) - 1

# Get the critical t-value
t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_freedom)

margin_of_error = t_critical * std_err_data

# Calculate confidence interval
confidence_interval = (mean_data - margin_of_error, mean_data + margin_of_error)

print(f"\n95% Confidence Interval for 'x.area_mean':")
print(f"  Sample Mean: {mean_data:.2f}")
print(f"  Margin of Error: {margin_of_error:.2f}")
print(f"  Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
print(f"Interpretation: We are {confidence_level*100}% confident that the true population mean of 'x.area_mean' lies within this interval.")
```

## Task 14: Analyse skewness and kurtosis to evaluate data distribution shape. Compute probability using conditional probability formula from dataset categories.

```python
# Analyse skewness and kurtosis for all numerical features

print("\nSkewness for numerical features:")
display(df[numerical_cols].skew())

print("\nKurtosis for numerical features:")
display(df[numerical_cols].kurtosis())

print("\nInterpretation of Skewness and Kurtosis:\nSkewness measures the asymmetry of the probability distribution. Positive skewness means the tail is on the right; negative skewness means the tail is on the left. Kurtosis measures the 'tailedness' of the distribution. High kurtosis indicates heavy tails (more outliers) and a sharper peak; low kurtosis indicates lighter tails and a flatter peak.")
```

```python
# Compute probability using conditional probability formula from dataset categories.
# Let's consider the 'y' column (diagnosis: M or B) and 'x.radius_mean' for an example.
# We will calculate P(y='M' | x.radius_mean > median_radius_mean)

median_radius_mean = df['x.radius_mean'].median()

# Event A: y == 'M' (Malignant)
event_a = df[df['y'] == 'M']

# Event B: x.radius_mean > median_radius_mean
event_b = df[df['x.radius_mean'] > median_radius_mean]

# Intersection of A and B: y == 'M' AND x.radius_mean > median_radius_mean
intersection_ab = event_a[event_a['x.radius_mean'] > median_radius_mean]

# Calculate P(A and B) = count(A and B) / total_count
p_a_and_b = len(intersection_ab) / len(df)

# Calculate P(B) = count(B) / total_count
p_b = len(event_b) / len(df)

# Calculate conditional probability P(A|B) = P(A and B) / P(B)
# Add a check to avoid division by zero
if p_b > 0:
    p_a_given_b = p_a_and_b / p_b
    print(f"\nConditional Probability P(y='M' | x.radius_mean > median_radius_mean):")
    print(f"  Median 'x.radius_mean': {median_radius_mean:.2f}")
    print(f"  P(y='M' and x.radius_mean > median_radius_mean): {p_a_and_b:.4f}")
    print(f"  P(x.radius_mean > median_radius_mean): {p_b:.4f}")
    print(f"  P(y='M' | x.radius_mean > median_radius_mean): {p_a_given_b:.4f}")
    print("Interpretation: This is the probability that a tumor is Malignant given that its mean radius is greater than the median mean radius.")
else:
    print("Cannot compute conditional probability: The event 'x.radius_mean > median_radius_mean' did not occur.")
```

## Task 18: Plot and compare Normal Distribution and Binomial Distribution curves.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom

# --- Normal Distribution ---
# Parameters for Normal Distribution (mean and standard deviation)
mu = 0  # Mean
sigma = 1 # Standard deviation

# Generate x-values for the normal distribution curve
x_normal = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)

# Calculate the Probability Density Function (PDF) values
pdf_normal = norm.pdf(x_normal, mu, sigma)

# --- Binomial Distribution ---
# Parameters for Binomial Distribution (number of trials, probability of success)
n = 10  # Number of trials
p = 0.5 # Probability of success

# Generate k-values (number of successes) for the binomial distribution
k_binomial = np.arange(0, n + 1)

# Calculate the Probability Mass Function (PMF) values
pmf_binomial = binom.pmf(k_binomial, n, p)

# --- Plotting ---
plt.figure(figsize=(12, 6))

# Plot Normal Distribution
plt.plot(x_normal, pdf_normal, label=f'Normal Distribution (μ={mu}, σ={sigma})', color='blue')

# Plot Binomial Distribution
plt.stem(k_binomial, pmf_binomial, label=f'Binomial Distribution (n={n}, p={p})', markerfmt='ro', linefmt='r-')

plt.title('Comparison of Normal and Binomial Distributions')
plt.xlabel('Value / Number of Successes')
plt.ylabel('Probability / Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(np.arange(min(k_binomial), max(k_binomial)+1, 1))
plt.tight_layout()
plt.show()
```
